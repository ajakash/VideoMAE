```python
import re
import matplotlib.pyplot as plt
```

```python
# From chatGPT 

# Function to extract numbers in the format "videos: xxxx"
def extract_numbers_from_file(file_path):
    numbers = []
    with open(file_path, 'r') as file:
        for line in file:
            match = re.search(r'videos:\s*(\d+)', line)
            if match:
                numbers.append(int(match.group(1)))
    return numbers

# File path to the text file
file_path = 'your_text_file.txt'

# Extract numbers
numbers = extract_numbers_from_file(file_path)

# Check if any numbers were found
if numbers:
    # Plot the extracted numbers
    plt.plot(numbers, marker='o', linestyle='-', color='b')
    plt.title('Numbers after "videos:"')
    plt.xlabel('Entry Index')
    plt.ylabel('Videos Count')
    plt.grid(True)
    plt.show()
else:
    print("No numbers found in the file.")
```
